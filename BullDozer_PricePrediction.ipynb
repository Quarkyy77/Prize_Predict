```python
from sklearn.metrics import mean_squared_log_error, mean_absolute_error, r2_score
X_train=df_train.drop("SalePrice",axis=1)
y_train=df_train['SalePrice']
X_valid=df_val.drop("SalePrice",axis=1)
y_valid=df_val['SalePrice']

def rmsle(ytest,ypred):
    return np.sqrt(mean_squared_log_error(ytest,ypred))

def show_scores(model):
    train_preds=model.predict(X_train)
    val_preds=model.predict(X_valid)
    scores={"TRAINING MAE ERROR":mean_absolute_error(y_train,train_preds),
            "VALIDATION MAE ERROR":mean_absolute_error(y_valid,val_preds),
           "ROOT MEAN SQUARED LOG ERROR TRAINING":rmsle(y_train,train_preds),
           "ROOT MEAN SQUARED LOG ERROR VALIDATION":rmsle(y_valid,val_preds),
           "R2 Score training=":r2_score(y_train,train_preds),
             "R2 score validation":r2_score(y_valid,val_preds)}
    return scores

     
    
```

```python
%%time
np.random.seed(42)
model=RandomForestRegressor(n_jobs=-1, max_samples=10000)
model.fit(X_train,y_train)
```

```python
show_scores(model)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
%%time
#USING RANDOMIZED SEARCH CV FOR HYPERPARAMETER TUNING}
grid={"n_estimators":np.arange(10,100,10),
     "max_depth":[None,3,5,10],
     "min_samples_split":np.arange(2,20,2),
     "min_samples_leaf":np.arange(1,20,2),
     "max_features":[0.5,1 ,'sqrt', "auto"],
     "max_samples":[10000]}
r_model=RandomizedSearchCV(RandomForestRegressor(n_jobs=-1, random_state=42),
                          param_distributions=grid,
                          n_iter=20,
                           cv=5,
                          verbose=True)
r_model.fit(X_train,y_train)
```

```python
r_model.best_params_
```

```python
show_scores(r_model)
```
